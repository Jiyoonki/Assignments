```python
import pandas as pd
```

# Read the data

```python
data = pd.read_csv("dedupe_challenge.csv")
```

```python
data = data.fillna('') # Changing 'Nan' values in order to do simpler comparison
```

```python
data.head(50)
```

## Simplest method

```python
df = data.drop_duplicates()
```

```python
df.head(50)
```

## Other alternative

```python
df2 = data.groupby(list(data.columns),sort=False).size().reset_index() # Grouping similar rows without sorting
df2 = df2.drop(0, axis =1)
```
